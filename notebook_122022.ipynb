### Nous sommes intéressés par:

#### 1.1)L'accès des résidents à Internet (pour l'apprentissage à distance) 
Pour évaluer l'accès de la population de chaque pays à l'apprentissage en ligne, nous pouvons utiliser l'indicateur "Internet users (per 100 people)". Cet indicateur mesure la proportion de la population de chaque pays qui peut accéder à l'apprentissage en ligne.
*'Internet users (per 100 people)''

#### 2)Le nombre d'etudiants potentiels par pays 
La proportion de la population ayant fait des études supérieures peut également être marquée 
car elle permet d'évaluer le niveau de la population jeune et la demande potentielle. 
*'Graduates from tertiary education, both sexes (number)'

#### 3)Croissance démographique 
Pour évaluer le potentiel des pays à vendre du contenu pour les étudiants à distance,
l'indicateur "Population, total" peut également être utile. 
Il permet d'estimer dans quels pays la population croît le plus intensément, 
ce qui peut indiquer une demande potentiellement croissante de services éducatifs.
*'Population, total'

#### 4)Solvabilité de la population.

Le PIB par habitant est un indicateur économique important 
qui peut indiquer le niveau de bien-être de la population et sa capacité de payer.
*'GDP per capita, PPP (current international $)'


```python
indicators = ['Population, total', 'GDP per capita, PPP (current international $)',
              'Internet users (per 100 people)',
              'Graduates from tertiary education, both sexes (number)']
```

```python
df = df.loc[df['Indicator Name'].isin(indicators), ['Country Name', 'Indicator Name', '2000', '2005', '2006', '2007', '2008','2009', '2010', '2011', '2012','2013']]
```

```python
df.head()
```

```python
gdp_pc = df[df['Indicator Name'] == 'GDP per capita, PPP (current international $)']
gdp_pc.head()
```

```python
fig, ax = plt.subplots()

for country in gdp_pc['Country Name']:
    ax.plot(gdp_pc.columns[2:], gdp_pc.loc[gdp_pc['Country Name'] == country].iloc[:,2:].values[0], label=country)
plt.title('GDP per capita, PPP (current international $ from 2000 to 2013)')
ax.set_xlabel('Years')
ax.set_ylabel('GDP per capita, PPP (current international $')
ax.legend(bbox_to_anchor=(1.05, 1.05), loc='upper left');
plt.savefig("my_plot7.JPEG", bbox_inches="tight")
```

```python
gdp_pc=gdp_pc.set_index("Country Name")
gdp_pc.drop(['Indicator Name'], axis=1, inplace=True)
```

```python
gdp_pc.groupby('Country Name').describe().transpose().head(15)
```

```python
gdp_pc['delta_gdp_pc']=(gdp_pc['2013'] - gdp_pc['2000']) 
gdp_pc['score_gdp_pc']=gdp_pc['delta_gdp_pc']/gdp_pc['delta_gdp_pc'].abs().max()

countries=gdp_pc.sort_values('delta_gdp_pc', ascending=False).iloc[:10]
countries
```

```python
inter_user=inter_user.set_index("Country Name")
inter_user.drop(['Indicator Name'], axis=1, inplace=True)
```

```python
sns.relplot(data=inter_user, x="2000", y="2013", hue="Country Name", kind="scatter");
plt.savefig("my_plot8.JPEG", bbox_inches="tight")
```

Pour créer des coefficients de notation pour chacun des indicateurs sélectionnés, il est nécessaire de les normaliser.
La normalisation nous permet d'amener toutes les valeurs des indicateurs 
dans la même plage de valeurs (de -1 à 1), ce qui nous permet de comparer objectivement différents indicateurs.

```python
inter_user['delta_inter_user']=inter_user['2013']-inter_user['2000']
inter_user['score_inter_user']=inter_user['delta_inter_user']/inter_user['delta_inter_user'].abs().max()
inter_user.sort_values('score_inter_user', ascending=False).iloc[:20]
```

```python
pop=df[df['Indicator Name']=='Population, total']
pop.head()
```

```python
pop=pop.set_index("Country Name")
pop.drop(['Indicator Name'], axis=1, inplace=True)
```

```python
pop.groupby('Country Name').agg(['mean', 'median', 'std'])
```

```python
pop['delta_population']=pop['2013']-pop['2000']
pop['score_population']=pop['delta_population']/pop['delta_population'].abs().max()
pop.sort_values('score_population', ascending=False).iloc[:10]
```

```python
top_10 = pop.sort_values('score_population', ascending=False).iloc[:20]

plt.figure(figsize=(8, 6))
plt.barh(top_10.index, top_10['delta_population'])
plt.title('Top 20  by Population Growth (2000-2013)')
plt.xlabel('Delta Population')
plt.ylabel('Country')

plt.savefig("my_plot11.JPEG", bbox_inches="tight")
```

```python
Graduat=df[df['Indicator Name']=='Graduates from tertiary education, both sexes (number)']

Graduat
```

```python
Graduat=Graduat.set_index("Country Name")
Graduat.drop(['Indicator Name'], axis=1, inplace=True)
```

```python
Graduat.groupby('Country Name').agg(['mean', 'median', 'std'])
```

```python
if (pop['delta_population'] == 0).any():
    print("delta_population value is 0")
else:
    print("delta_population value is not  0")   
```

```python
Graduat['delta_Graduates']=(Graduat['2013']-Graduat['2000'])/pop['delta_population']
Graduat['score_Graduates']=Graduat['delta_Graduates']/Graduat['delta_Graduates'].abs().max()
Graduat.sort_values('score_Graduates', ascending=False).iloc[:15]
```

```python

top_15 = Graduat.sort_values('score_Graduates', ascending=False).iloc[:15]

plt.figure(figsize=(8, 6))
plt.barh(top_15.index, top_15['delta_Graduates'])
plt.title('Top 15 Countries by Graduates Growth per Delta Population (2000-2013)')
plt.xlabel('Delta Graduates')
plt.ylabel('Country')
plt.savefig("my_plot12.JPEG", bbox_inches="tight")

```

### Donnerons la liste des 20 pays les mieux classés par rapport à `score_synt`

```python
score_synt=(gdp_pc['score_gdp_pc']+pop['score_population']+Graduat['score_Graduates']+inter_user['score_inter_user'])/4
score_synt.sort_values(ascending=False).iloc[:20]
```

```python
data_Test = pd.merge(gdp_pc['score_gdp_pc'], pop['score_population'], on='Country Name')
data_Test
```

```python
data_Test = pd.merge(Graduat['score_Graduates'],data_Test, on='Country Name')
data_Test
```

```python
data_Test = pd.merge(data_Test, inter_user['score_inter_user'], on='Country Name')
data_Test
```

```python
data_Test['score_synt']=data_Test.mean(axis=1)
```

```python
data_Test
```

```python

data_Test.groupby('Country Name')['score_synt'].sum().sort_values(ascending=False).iloc[:20]
```

```python
corr_matrix = data_Test.corr()
corr_matrix
```

```python
plt.figure(figsize=(10,8))
sns.set(font_scale=1.2)
sns.set_style('white')
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True, annot_kws={'size':12})
plt.title('Heatmap corr')
plt.savefig("my_plot.JPEG", bbox_inches="tight")
```

```python

plt.figure(figsize=(12, 10))

mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

sns.heatmap(corr_matrix, mask=mask, center=0, cmap='RdBu', linewidths=1, annot=True, annot_kws={'size':8})

plt.title('Coefficient de corrélation de Pearson entre variables', fontweight='bold')
plt.savefig("my_plot.JPEG", bbox_inches="tight")
```
